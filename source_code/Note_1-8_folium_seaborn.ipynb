```python
import folium
```

```python
# .Map(location=[위경도],zoom_start = 13) zoom_start : 초기 확대 비율 정의
map_osm = folium.Map(location = [45.5236, - 122.6750], zoom_start= 13)
map_osm
```

```python
# 지도 스타일 변경 : tiles = Stamen Toner/Stamen Terrain
map_osm = folium.Map(location = [45.5236, - 122.6750], zoom_start= 13, tiles='Stamen Toner')
map_osm
```

```python
map_1 = folium.Map(location = [45.5236, - 122.6750], zoom_start= 13, tiles='Stamen Toner')
folium.Marker([45.5236, - 122.6750], popup = 'Mt. Hood Meadowd', icon= folium.Icon(icon='cloud',color = 'red')).add_to(map_1)
map_1
```

```python
# 특정위치를 기준으로  지도를 표시하고 마커 생성해서 찍어볼 것
# 현재 있는 곳의 위경도를 파악해서 지도를 표시하고 마커 생성해서 찍어볼 것
# 37.507923394520084, 127.03015610685301

map_2 = folium.Map(location = [37.507923394520084, 127.03015610685301], zoom_start= 16)
folium.Marker([37.507923394520084, 127.03015610685301], popup = 'Mt. Hood Meadowd', icon= folium.Icon(icon='cloud',color = 'red')).add_to(map_2)
map_2
```
