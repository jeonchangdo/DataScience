### 열/행 인덱스 변환(교환) 함수 : stack()
- 열인덱스 -> 행인덱스로 변환

### 열/행 인덱스 변환(교환) 함수 : unstack()
- 행인덱스 -> 열인덱스로 변환

### df.stack(변환할 열 인덱스)

### df.unstack(변환할 행 인덱스)
- 두 함수 모두 원본 방영 안됨 - 필요시 저장

```python
df
df.stack('Cidx1') # 열 > 행 인덱스 변환
df # 원본 반영 안됨
```

```python
df.stack('Cidx1').unstack('Cidx1')
```

```python
np.random.seed(0)
df1 = pd.DataFrame(np.round(np.random.randn(6,4),2),
                  columns=[['A', 'A', 'B', 'B'],
                           ['C1', 'C2', 'C1', 'C2']],
                 index = [['M','M','M','F','F','F'],
                         ['id_' + str(i) for i in range(1,4)]*2]) 
df1
```

```python
df1.columns.names = ['Cidx1', 'Cidx2']
df1.index.names = ['Ridx1', 'Ridx2']
df1
```

```python
# df1 열 인덱스 cidx1을 행 인덱스로 변환하여
# 출력하시오
df1.stack('Cidx1')
```

```python
# stack() 함수 사용시 컬럼 name 대신 level 값을 사용해도 됨
df1.stack(0)
# stack(열 제목 또는 열 위치 level 값)
```

```python
# unstack() : 행인덱스 > 열인덱스로 변환 : 원본 반영 안됨
df1
df1.unstack('Ridx2')
df1.unstack(0)
df1.unstack(0).B.C1.F
```

### 다중인덱스의 접근 방법 정리
- 인덱스가 하나가 아니기 때문에 묶어서(튜플로) 전달해야한다.
- df[(튜플)] - 열 접근
- df.loc[(튜플)] - 행 접근

```python
df1
```

```python
# df1 데이터 프레임을 사용해서 첫번째 열과 첫번째 행을 출력
# 첫번째 열
df1.A.C1
df1[('A','C1')]
```

```python
# 첫번째 행
df1
df1.loc[('M','id_1')]
# df1.loc.M.id_1 # loc 인덱서는 '.' 연산자 사용불가
```

```python
# 다중 인덱싱의 원소 접근 방법
# loc 인덱서를 사용하는 것이 가장 좋다
df1
df1.loc[('M','id_1'),('A','C1')]

# 4행 4열의 원소값 0.33일 loc 인덱서로 접근해서 출력하시오.
df1.loc[('F','id_1'),('B','C2')]

# 4행 4열의 원소를 접근해서 값을 33으로 변경하시오. loc 인덱서 사용할 것
df1.loc[('F', 'id_1'),('B','C2')] = 33
df1
```

```python
# iloc 인덱서 사용
# df1이 다중 인덱싱을 하고 있지만 iloc는 위치 접근이기 때문에 인덱스에 구애 받지 않는다.
df1
df1.iloc[3,3]
```

```python
# M의 id_1 행의 A 열을 반환하시오.
df1.loc[('M','id_1'),'A']
```

```python
# F행을 모두 반환하시오.
df1.loc['F']
```

```python
# 데이터 프레임에 계산 필드 추가
# 각 열의 끝행과 끝열에 행과 열의 총 합을 추가 하시오.
```

```python
np.random.seed(0)
df1 = pd.DataFrame(np.round(np.random.randn(6,4),2),
                  columns=[['A', 'A', 'B', 'B'],
                           ['C1', 'C2', 'C3', 'C4']],
                 index = [['M','M','M','F','F','F'],
                         ['id_' + str(i) for i in range(1,4)]*2]) 
df1
```

```python
# 열방향 총 합 계산

df1.loc['All'] = df1.sum() # 행 인덱스 처리하지 않아서 인덱스 구조가 변경
df1
```

```python
df1.loc[('All','All')] = df1.sum() # 행과 열 모두 새우이 인덱스를 생성하고 계산이 발생하지 않는다
df1
```

```python
df1.loc[('All','All'),:] = df1.sum()
df1
df1[('All','All')] = df1.sum(axis=1)
df1
```

```python
# 다중 인덱스 순서 교환
# swaplevel(i,j,axis=0/1) 0 : 행(디폴트) 1: 열
# i와 j는 교환하고자 하는 인덱스 라벨, 위치번호
```

```python
df2 = df1.swaplevel(0,1,0) # 행 인덱스의 레벨 교환
df2
```

```python
df2 = df1.swaplevel(0,1,1) # 행 인덱스의 레벨 교환
df2
```
